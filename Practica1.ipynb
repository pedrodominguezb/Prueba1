```python
lista_fill=[]
for i in range(0,20):
 lista_fill.append(i)
lista_fill
```

```python
index=0
while index<len(lista_fill):
  print(lista_fill[index])
  index +=1
```

```python
intervalo=4
contador=0
split_list=[]

while contador<len(lista_fill):
  split_list.append(lista_fill[contador:contador+intervalo])
  contador +=intervalo
print (split_list)
```
